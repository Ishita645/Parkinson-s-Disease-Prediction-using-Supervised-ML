## Classifying the dataset into features and labels

```python
X1=dataset.drop(columns=['name','status'], axis=1)
X=X1.iloc[:, :].values
#y1=dataset['status']
y=dataset.iloc[:,17].values
```

## Spliting the dataset into Training and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

```python
print(X_train)
```

```python
print(X_train)
```

## Standardization

#### Standardization is used to scale down the values of features and labels to values between -3 to +3 so that we obtain values in a range 

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

#### fit method is used to calculate the mean and variance

```python
#scaler.fit(X_train)
```

#### Transform is used to turn all values of features according to standardised/fit values

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_test.shape
```

```python
print(X_train)
```

 ## Training the Support Vector machine model into the Training and Test Set

```python
from sklearn.svm import SVC
model = SVC(kernel='linear',random_state = 0)
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(y_train, X_train_prediction)
```

```python
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(y_test, X_test_prediction)
```

```python
print('Accuracy score of test data : ', test_data_accuracy)
```
